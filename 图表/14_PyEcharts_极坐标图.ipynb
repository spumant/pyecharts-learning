```python
from pyecharts import options as opts
from pyecharts.charts import Polar

import random
from pyecharts.faker import Faker
```

```python
data = [(i, random.randint(1, 100)) for i in range(101)]
data
```

```python
c = (
    Polar()
    .add(
        '极坐标',
        data,
        type_='scatter',  # 散点图
        label_opts=opts.LabelOpts(is_show=False)
    )
    .set_global_opts(title_opts=opts.TitleOpts(title='极坐标图'))
)
c.render_notebook()
```

```python
Faker.week
```

```python
c = (
    Polar()
    .add_schema(
        radiusaxis_opts=opts.RadiusAxisOpts(
            data=Faker.week,
            type_='category'
        ),
        angleaxis_opts=opts.AngleAxisOpts(
            is_clockwise=True,
            max_=10
        )
    )
    .add('商品', [1, 2, 3, 4, 3, 5, 1], type_='bar')
    .set_global_opts(title_opts=opts.TitleOpts(title='极坐标图'))
)
c.render_notebook()
```

```python
c = (
    Polar()
    .add_schema(
#         radiusaxis_opts=opts.RadiusAxisOpts(
#             data=Faker.week,
#             type_='category'
#         ),
        angleaxis_opts=opts.AngleAxisOpts(
            data=Faker.week,
            type_='category'
        ),
    )
    .add('商品A', [1, 2, 3, 4, 3, 5, 1], type_='bar', stack='abc')
    .add('商品B', [2, 4, 3, 5, 2, 3, 1], type_='bar', stack='abc')
    .add('商品C', [1, 2, 3, 4, 3, 5, 1], type_='bar', stack='abc')
    .set_global_opts(title_opts=opts.TitleOpts(title='极坐标图+堆叠柱形图'))
)
c.render_notebook()
```

```python
data = [(i, random.randint(1, 100)) for i in range(10)]
```

```python
c = (
    Polar()
    .add(
        '极坐标',
        data,
        type_='effectScatter',  # 涟漪散点图
        effect_opts=opts.EffectOpts(
            scale=10, # 涟漪范围大小
            period=1,  # 动画周期
        )
    )
    .set_global_opts(title_opts=opts.TitleOpts(title='极坐标图'))
)
c.render_notebook()
```
