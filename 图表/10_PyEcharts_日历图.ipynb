```python
from pyecharts import options as opts
from pyecharts.charts import Calendar

import datetime
import random
```

```python
# 2023å¹´
begin = datetime.date(2023, 1, 1)
end = datetime.date(2023, 12, 31)
```

```python
(end - begin).days
```

```python
data = [ [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
        for i in range((end - begin).days + 1)]
```

```python
data
```

```python
c = (
    Calendar()
    .add(
        '',
        data,
        calendar_opts=opts.CalendarOpts(
            range_='2023',
            daylabel_opts=opts.CalendarDayLabelOpts(name_map='cn'),  # ä¸­æ
            monthlabel_opts=opts.CalendarMonthLabelOpts(name_map='cn'),  # ä¸­æ
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='2023å¹´æ¥åå¾'),
        
        visualmap_opts=opts.VisualMapOpts(
            max_=25000,
            min_=1000,
            orient='horizontal',  # æ°´å¹³æ¹å
            is_piecewise=True,  # æ¾ç¤ºæ¹å¼
            pos_left='100px',
            pos_top='230px'
        )
    )
)
c.render_notebook()
```
