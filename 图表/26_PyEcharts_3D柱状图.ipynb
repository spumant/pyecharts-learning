```python
from pyecharts import options as opts
from pyecharts.charts import Bar3D

from pyecharts.faker import Faker
import random
```

```python
data = [(i, j, random.randint(0, 12)) for i in range(6) for j in range(24)]
data
```

```python
c = (
    Bar3D()
    .add(
        '',
        [[d[1], d[0], d[2]]  for d in data],
        xaxis3d_opts=opts.Axis3DOpts(Faker.clock, type_='category'),
        yaxis3d_opts=opts.Axis3DOpts(Faker.week, type_='category'),
        zaxis3d_opts=opts.Axis3DOpts(type_='value')
    )
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(max_=15)
    )
)
c.render_notebook()
```

### 堆叠3D柱状图

```python
x_data = y_data = list(range(10))
```

```python
x_data, y_data
```

```python
def generate_data():
    
    data = []

    for j in range(10):
        for k in range(10):
            value = random.randint(0, 9)
            data.append([j, k, value*2 + 4])

    return data
```

```python
generate_data()
```

```python
bar3d = Bar3D()

for _ in range(10):
    
    bar3d.add(
        '',
        generate_data(),
        shading='lambert',  # 更清晰，阴影效果
        
        xaxis3d_opts=opts.Axis3DOpts(x_data, type_='value'),
        yaxis3d_opts=opts.Axis3DOpts(y_data, type_='value'),
        zaxis3d_opts=opts.Axis3DOpts(type_='value')
    
    )

bar3d.set_global_opts(title_opts=opts.TitleOpts('3D堆叠柱状图'))
bar3d.set_series_opts(stack='abc')

bar3d.render_notebook()
```
