```python
from pyecharts import options as opts
from pyecharts.charts import Tree
```

```python
data = [
    {
        'children':[
            {'name': '叔叔'},
            {
                'children': [{'name': '小红'}],
                'name': '伯伯'
            },
            {
                'name': '父亲',
                'children': [
                    {'name': '我'},
                    {'name': '姐姐'}
                ]
            }
            
        ],
        'name': "祖父母"
    }
]
```

```python
c = (
    Tree()
    .add('', data)
    .set_global_opts(
        title_opts=opts.TitleOpts(title='树图')
    )
)
c.render_notebook()
```
