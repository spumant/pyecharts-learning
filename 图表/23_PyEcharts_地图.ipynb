```python
from pyecharts import options as opts
from pyecharts.charts import Map

from pyecharts.faker import Faker
```

### 地图

```python
Faker.provinces
```

```python
provinces = [p + "市" if p=='上海' else p + "省" for p in Faker.provinces]
provinces
```

```python
c = (
    Map()
    .add(
        '',
        [list(i) for i in zip(provinces, Faker.values())],
        'china'  # 地图类型
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='地图'),
        visualmap_opts=opts.VisualMapOpts(
            max_=200,
            is_piecewise=True
        )
    )
)
c.render_notebook()
```

### 世界地图

```python
c = (
    Map()
    .add(
        '',
        [list(i) for i in zip(Faker.country, Faker.values())],
        'world'  # 地图类型
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False)
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='世界地图'),
        visualmap_opts=opts.VisualMapOpts()
    )
)
c.render_notebook()
```

### 地球

```python
from pyecharts.charts import MapGlobe
from pyecharts.faker import POPULATION
```

```python
POPULATION
```

```python
# 所有人口数据
POPULATION[1:]
```

```python
data = [x for _,x in POPULATION[1:]]
data  # 人口
```

```python
low, high = min(data), max(data)
```

```python
low, high
```

```python
c = (
    MapGlobe()
    .add_schema()
    .add(
        maptype='world',
        series_name='世界人口',
        data_pair=POPULATION[1:], # 人口数据
        
        is_map_symbol_show=False, # 地图中各个国家的符号是否显示
        label_opts=opts.LabelOpts(is_show=False)  # 显示国家名称
    )
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            min_=low,
            max_=high,
            range_text=['Max', 'Min'],
            is_calculable=True,
            range_color=['lightskyblue', 'yellow', 'orange']
        )
    )
)
c.render_notebook()
```
