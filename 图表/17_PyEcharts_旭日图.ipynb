```python
from pyecharts.charts import Sunburst
from pyecharts import options as opts
```

```python
data = [
    # 旭日项
    opts.SunburstItem(
        name="Grandpa",
        children=[
            opts.SunburstItem(
                name="Uncle",
                value=15,
                children=[
                    opts.SunburstItem(name="Jack", value=2),
                    opts.SunburstItem(
                        name="Mary",
                        value=5,
                        children=[opts.SunburstItem(name="Jackson", value=2)],
                    ),
                    opts.SunburstItem(name="Ben", value=4),
                ],
            ),
            opts.SunburstItem(
                name="Father",
                value=10,
                children=[
                    opts.SunburstItem(name="Me", value=5),
                    opts.SunburstItem(name="Peter", value=1),
                ],
            ),
        ],
    ),
    opts.SunburstItem(
        name="Nancy",
        children=[
            opts.SunburstItem(
                name="Nike",
                children=[
                    opts.SunburstItem(name="Betty", value=1),
                    opts.SunburstItem(name="Jenny", value=2),
                ],
            )
        ],
    ),
]
```

```python
c = (
    Sunburst(
        init_opts=opts.InitOpts(
            width='800px', height='400px'
        )
    )
    .add('', data_pair=data)
    .set_global_opts(title_opts=opts.TitleOpts(title='旭日图'))
)
c.render_notebook()
```
