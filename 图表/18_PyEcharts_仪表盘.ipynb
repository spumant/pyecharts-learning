```python
# 仪表盘
from pyecharts.charts import Gauge
import pyecharts.options as opts
```

```python
c = Gauge()
c.add('指标', data_pair=[('', '66')], radius='60%')
c.render_notebook()
```

```python
# 链式
c = (
    Gauge()
    .add('销售额完成率', [('', 60)])
    .set_global_opts(title_opts=opts.TitleOpts(title='仪表盘'))
)
c.render_notebook()
```
